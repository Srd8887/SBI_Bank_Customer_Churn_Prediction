### Why use stratify=y?

- If your dataset has imbalanced classes, e.g.:
*    Exited = 0 → 80%
*    Exited = 1 → 20%

- Then using stratify=y will maintain the same 80:20 ratio in both x_train/y_train and x_test/y_test

###  stratify=y in train_test_split means:

*  It ensures that the proportion of classes in the y (target) variable is the same in both the training and test sets.

### Without stratify:

*    The class distribution may become skewed in training/test sets (like all 0s in test set!), leading to poor model evaluation.

# Handling Imbalanced Data with Smote

# SMOTE (Synthetic Minority Over-sampling Technique)

- SMOTE is used to solve the class imbalance problem in classification tasks — especially when one class (e.g., 1) is rare compared to the other (e.g., 0).

  # Why Use SMOTE?

  - Because Logistic Regression (and most classifiers) tend to favor the majority class, leading to:

  -  Poor recall for the minority class

   -  High overall accuracy but poor performance on rare outcomes



| Class | Count |
| ----- | ----- |
| 0     | 900   |
| 1     | 100   |


```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Split data FIRST
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Apply SMOTE only on the training data
sm = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = sm.fit_resample(X_train, y_train)

# Now train your model on the resampled data
model = LogisticRegression()
model.fit(X_train_resampled, y_train_resampled)

# Predict only on the original test data
y_pred = model.predict(X_test)
```

```python
# Now this will work (same size)
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,classification_report
print(confusion_matrix(y_test,y_test))
print(pd.crosstab(y_test,y_pred))
print("🔍 Accuracy:", accuracy_score(y_test, y_pred))
print("📊 Precision:", precision_score(y_test, y_pred))
print("📈 Recall:", recall_score(y_test, y_pred))
print("🎯 F1 Score:", f1_score(y_test, y_pred))
print("\n📄 Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from imblearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, StratifiedKFold
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score

# Set up pipeline
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('model', LogisticRegression(max_iter=1000))
])

# Stratified k-fold (preserves class ratio)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Run CV and evaluate multiple scores
scoring_metrics = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score)
}

# Run CV scores for all metrics
from sklearn.model_selection import cross_validate

cv_results = cross_validate(pipeline, X, y, cv=cv, scoring=scoring_metrics)

# Display results
print("📊 Cross-Validation Results:")
for metric in scoring_metrics:
    print(f"{metric.capitalize()} (Mean): {cv_results['test_' + metric].mean():.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# 2. Apply SMOTE to training data only
sm = SMOTE(random_state=42)
X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train)

# 3. Train Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_smote, y_train_smote)

# 4. Predict on original test data
y_pred = model.predict(X_test)

# 5. Evaluate
print("\n🧪 Final Test Set Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# 6. Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Blues')
plt.title("Confusion Matrix")
plt.grid(False)
plt.show()

```
